# Creating Feature Vectors


```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
tokenized_features = [word_tokenize(x) for x in data2["features"]]
```

```python
import gensim
model = gensim.models.Word2Vec(tokenized_features, vector_size=300, window=5, min_count=1, workers=8)
# Access the word vectors
word_vectors = model.wv
```

```python
place_review_array = []
for i in range(len(data2['features'])):
  word_vector = np.zeros((1,300))
  for word in word_tokenize(data2['features'].iloc[i]):
    
    word_vector += word_vectors[word]
  word_vector = word_vector/len(word_tokenize(data2['features'].iloc[i]))
  place_review_array.append(word_vector) 
  
```

```python
place_review_array = np.array(place_review_array)
place_review_array = place_review_array.reshape(-1,300)
place_review_array.shape
```

# Recommendation System (Content Based)

```python
def Recommend():
  user_review = str(input("Enter your query: "))
  print("\n")
  user_review = user_review.lower()
  user_review_array = np.zeros((1, 300))
  for word in word_tokenize(user_review):
    if word in word_vectors:
      user_review_array += word_vectors[word]
    
  user_review_array = user_review_array/len(user_review)
  similarities = cosine_similarity(place_review_array,user_review_array).flatten()
  data2['Similarity'] = similarities
  df=data2.sort_values('Similarity',ascending=False)
  df = df.drop_duplicates(subset=['name'])
  return df
```

```python
import warnings
warnings.filterwarnings('ignore')

df = Recommend()
display(df.head(5))
```
